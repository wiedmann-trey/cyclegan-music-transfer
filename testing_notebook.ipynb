```python
genre_dictionary = {}

def create_genre_dictionary(root):
    for dirpath, dirs, filenames in os.walk(root):
        for file in filenames:

            absolute_file = os.path.join(root, file)

            with open(absolute_file, "r", encoding="utf-8", errors="ignore") as opening_file: 
                lines = opening_file.read().split()

                for i in lines: 
                    genre_dictionary[i] = file[:-4] #get everything up until .txt
                    
    #this was just to for me to see what it looked like lolol        
    with open("/Users/carolinezhang/Downloads/cyclegan-music-transfer/song_dictionary.txt", "w") as output:
        output.write(str(genre_dictionary))
    

create_genre_dictionary("/Users/carolinezhang/Downloads/cyclegan-music-transfer/labels for songs")
print(genre_dictionary)

```
