```python
genre_dictionary = {}

def create_genre_dictionary(root):
    for dirpath, dirs, filenames in os.walk(root):
        for file in filenames:

            absolute_file = os.path.join(root, file)

            with open(absolute_file, "r", encoding="utf-8", errors="ignore") as opening_file: 
                lines = opening_file.read().split()

                for i in lines: 
                    genre_dictionary[i] = file[:-4] #get everything up until .txt
                    
    #this was just to for me to see what it looked like lolol        
    with open("song_dictionary.txt", "w") as output:
        output.write(str(genre_dictionary))
    

create_genre_dictionary("labels for songs")
print(genre_dictionary)

```

```python
import os
import shutil

song_paths = []
song_genre_dict = {}
def get_genre_path(root):
    
    for dirpath, dirs, filenames in os.walk(root):
        for dir in dirs:
            if len((dir)) >= 5:
                song_paths.append(os.path.join(dirpath, dir))
                try: 
                    song_genre = genre_dictionary[dir]
                except KeyError as e:
                    genre_dictionary[dir] = "unknown"
                song_genre_dict[os.path.join(dirpath, dir)] = dir
    # just for me to see lolol           
    with open("song_paths.txt", "w") as output:
        output.write(str(song_paths))
    return song_paths, song_genre_dict

print(get_genre_path("lmd_matched"))
```

```python
def create_genre_number_dict(root):
    """creates a dictionary mapping all the string labels to a number which we can later use for one hot encoding"""
    genre_numerical_labels = {}
    for dirpath, dirs, filenames in os.walk(root):
        n = 0
        for file in filenames:
            if file[0] != ".": #cuz for some reason theres a .dsstore in labels for songs but idk how it got there
                genre_numerical_labels[file[:-4]] = n
                n += 1
    genre_numerical_labels["classical"] = len(genre_numerical_labels) #had to manually put clasical in cuz yamahas all classical but no classical in lakh
    return genre_numerical_labels

genre_number_dict = create_genre_number_dict("labels for songs")
print(genre_number_dict)
```

```python
# I'm not really sure what adjust resolution means in the MusPy documentation. For the key change one though that's specific to chords not music objects
def add_augmentation(muspy_object):
    if np.random.random() < 0.1:
        return muspy_object.adjust_resolution(target=1, factor=0.1)
    return muspy_object
```

```python
muspy_objects = []
invalid_midis = []
muspy_labels = []
def add_muspy_objects():
    for song_path in song_paths:
        for dirpath, dirs, midifiles in os.walk(song_path):
            for midifile in midifiles: 
                absolute_song_path = os.path.join(song_path, midifile)
                try: 
                    muspy_obj = muspy.read(absolute_song_path)
                    muspy_obj = add_augmentation(muspy_obj)
                    just_song_key = song_genre_dict[song_path]
                    song_genre = genre_dictionary[just_song_key]
                    muspy_obj.annotations = song_genre
                    muspy_obj = muspy.to_event_representation(music=muspy_obj, encode_velocity=True, max_time_shift=100, velocity_bins=32)
                    muspy_objects.append(muspy_obj)
                    genre_number = genre_number_dict[song_genre]
                    muspy_labels.append(genre_number)
                    print(muspy_labels)
                except Exception as e:
                    print(repr(absolute_song_path))
                    invalid_midis.append(absolute_song_path)
                    continue
                
            
def remove_invalid_midi():
    os.mkdir("invalid_midis")
    for invalid_midi in invalid_midis:
        shutil.move(invalid_midi, "invalid_midis")
        
add_muspy_objects()
#remove_invalid_midi()
```

```python

def clean_yamaha(root):
    for dirpath, dirs, filenames in os.walk(root):
        for filename in filenames: 
            #if filename.endswith(".midi") or filename.endswith(".mid"):
                rel_song_path = os.path.join(dirpath, filename)
                try: 
                    new_yamaha = muspy.read(rel_song_path)
                    print(rel_song_path)
                    new_yamaha = add_augmentation(new_yamaha)
                    new_yamaha = muspy.to_event_representation(music=new_yamaha, encode_velocity=True, max_time_shift=100, velocity_bins=32)
                    muspy_objects.append(new_yamaha) 
                    song_genre = "classical"
                    genre_number = genre_number_dict[song_genre]
                    muspy_labels.append(genre_number)
                    print("not corrupted!")
                except Exception as e:
                    print(repr(filename))
                    continue
    with open("yamaha_paths.txt", "w") as output:
        output.write(str(muspy_objects))

clean_yamaha("maestro-v3.0.0")
```
